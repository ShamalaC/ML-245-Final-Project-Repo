```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['MLP Classifier', 'LSTM RNN', 'Decision Tree', 'SGD Classifier', 'XGBoost', 
          'Naive Bayes', 'KNN', 'Random Forest', 'CatBoost Classifier']
accuracies = [0.75, 0.72, 0.9, 0.65, 0.82, 0.55, 0.88, 0.92, 0.8]

# Plotting the bar chart
fig, ax = plt.subplots()

bars = ax.bar(models, accuracies, color='skyblue')

# Add labels, title, and customize the x-axis labels
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Comparison of Model Accuracies')
ax.set_xticklabels(models, rotation=45, ha='right')

# Customize the y-axis to range from 0 to 1
ax.set_ylim(0.5, 1.0)

# Display the values on top of each bar
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['XGBoost', 'Naive Bayes', 'KNN', 'CatBoost', 'Decision Tree', 
          'Random Forest', 'RNN (LSTM)', 'SGD Classifier', 'MLP Classifier']
precisions = [0.83, 0.63, 0.89, 0.80, 0.89, 0.91, 0.73, 0.70, 0.73]

# Plotting the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 5))

bars = ax.barh(models, precisions, color='purple')

# Add labels, title, and customize the y-axis labels
ax.set_ylabel('Model')
ax.set_xlabel('Precision')
ax.set_title('Precision vs Models')

# Display the values on the bars
for bar in bars:
    width = bar.get_width()
    ax.annotate(f'{width:.2f}',
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(3, 0),  # 3 points horizontal offset
                textcoords="offset points",
                ha='left', va='center')

# Invert y-axis to have the highest values at the top
ax.invert_yaxis()

# Show the plot
plt.tight_layout()
plt.show()
```
